```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
import numpy as np
# load data into a pandas DataFrame
data = pd.read_csv('Salary_Data.csv')
# separate independent variables (age, experience) 
# from dependent variable (salary)
X = data[['Age', 'YearsExperience']]
y = data['Salary']
# create a linear regression model and fit it to the data
model = LinearRegression()
model.fit(X, y)
# plot the data and the linear regression plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# plot the linear regression plane as a wireframe, colored in yellow
x1_grid, x2_grid = np.meshgrid(np.arange(X['Age'].min(), X['Age'].max(), 0.1), 
np.arange(X['YearsExperience'].min(), X['YearsExperience'].max(), 0.1))
y_grid = model.intercept_ + model.coef_[0] * x1_grid + model.coef_[1] * x2_grid
ax.plot_wireframe(x1_grid, x2_grid, y_grid, rstride=1, cstride=1, color='blue', 
alpha=0.05)
# plot the data points as a scatter plot, colored in red
ax.scatter(X['Age'], X['YearsExperience'], y, c='red', marker='o')
ax.set_xlabel('Age')
ax.set_ylabel('Years of Experience')
ax.set_zlabel('Salary')
plt.show()
```
