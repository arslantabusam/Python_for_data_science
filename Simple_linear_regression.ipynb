```python
#Simple Linear Regression
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.linear_model import LinearRegression
x = np.array([49, 57, 61, 68, 69, 76, 78])
y = np.array([122, 135, 147, 146, 162, 155, 163])
plt.scatter(x, y, color="blue");
#creates an instance of the LinearRegression class
model = LinearRegression(fit_intercept=True)
#the following code fits the model to the data
#np.newaxis converts the 1D x array to a 2D array with a single column
model.fit(x[:, np.newaxis], y)
#creates an array of 1000 evenly spaced numbers between 45 and 80
x_fit = np.linspace(45, 80, 1000)
#predicts the y-values for the input values in xfit
y_fit = model.predict(x_fit[:, np.newaxis])
plt.scatter(x, y)
plt.plot(x_fit, y_fit, color="red");
plt.xlabel('\n Height (inches) \n ')
plt.ylabel('\nWeight (pounds)\n')
plt.show()
```
